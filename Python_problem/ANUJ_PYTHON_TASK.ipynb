# IMPORTING THE REQUIRED LIBRARIES

```python
from tweepy import OAuthHandler
from tweepy import API
import json
import pandas as pd
```

# SETTING THE VALUES OF THE KEYS NEEDED TO USE TWEEPY

```python
consumer_key = "XXXXXXXXXXXXXXXXXXXXXXXXXXXX" 
consumer_secret = "XXXXXXXXXXXXXXXXXXXXXXXXXX"
access_key = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
access_secret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
```

# MAKING THE OBJECT OF THE API BY FIRST VERIFYING THE ACCOUNT

```python
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
auth_api = API(auth)
```

# FUNCTION TO GET THE TWEETS AND THEIR INFORMATION AND SAVE IT TO A FILE

```python
#Function for getting maximum permissible tweets and saving it to json file
def saveTweetsToJSON(screen_name):
    with open(screen_name+".jsonl",'w') as fp:
        for status in auth_api.user_timeline(screen_name=screen_name,count=200):
            fp.write(json.dumps(status._json)+"\n")
```

# CALLING THE FUNCTION WITH ARGUMENT AS "midasIIITD"

```python
name=input("ENTER SCREEN NAME OF A TWITTER ACCOUNT: ")
saveTweetsToJSON(name)
```

# USING THE PANDAS LIBRARY TO GET A TABULAR FORMAT OF THE DATA

```python
df=pd.read_json(name+".jsonl",lines=True)
```
